## 5.21.3 Mengurutkan data pada Pandas

```python
#FUNGSI SORT_VALUES() digunakan untuk melakukan pengurutan berdasarkan dengan kolom yang disebutkan.
#Data yang telah dipanggil dapat diurutkan mulai dari tinggi ke rendah ataupun dari rendah ke tinggi untuk data numerik
#Data stirng akan diurutkan berdasarkan abjad.
```

```python
#Mengurutkan data tabel nama berdasarkan abjad a sampai z
print (dt.sort_values ('Nama') .head ()) #tanpa parameter, akan menampilkan data 5 baris pertama
print (dt.sort_values ('Nama') .head (3))  #dengan parameter 3, menampilkan data 3 baris pertama

#Mengurutkan data tabel pedal width (cm) berdasarkan nilai terkecil ke terbesar
print (df.sort_values ('petal width (cm)').head()) #tanpa parameter, akan menampilkan 5 baris pertama
print (df.sort_values ('petal length (cm)') .head (4))  #dengan parameter 4,menampilkan data 4 baris pertama

#Mengurutkan data tabel petal length (cm) berdasarkan nilai terkecil ke terbesar
print (dt.sort_values ('Nama') .tail ())  #tanpa parameter, akan menampilkan 5 baris terakhir
print (dt.sort_values ('Nama') .tail (2)) #dengan parameter 2, akan menampilkan 2 baris terakhir
print (df.sort_values ('petal length (cm)') .tail ())#tanpa parameter, akan menampilkan 5 baris terakhir
print (df.sort_values ('petal length (cm)') .tail (2)) #dengan parameter 2, akan menampilkan 2 baris terakhir
```

```python
 #Untuk mengurutkan data berdasarkan abjad, dari a sampai z, maka parameter ascending =true
print (dt.sort_values ('Nama',ascending=True))
print (df.sort_values ('petal length (cm)',ascending=True))

#Untuk mengurutkan data berdasarkan abjad, dari z sampai a, maka parameter ascending = false
print (dt.sort_values ('Nama',ascending=False))
print (df.sort_values ('petal length (cm)',ascending=False))
```

## 5.22.4 Properti Pembuatan Histogram dengan Matplotlib

```python
# Mendefinisikan data untuk kumpulan data yang akan diproses pada diagram batang
# Penyanyi sebagai label pada sumbu x, dan penggemar sebagai nilai pada sumbu y
penyanyi=['Armand','Ariel','Once','Ari Lasso']
penggemar=[99,49,29,150]

# Mengatur ukuran dari figure yang akan menampilkan histogram.
plt.figure (figsize=(8,4))

# Membuat histogram dengan memasukkan penyanyi sebagai label pada sumbu x dan penggemar sebagai nilai pada sumbu y.
histogram=plt.bar(penyanyi, penggemar, color='yellow')

# Menentukan pola pada setiap bar dari histogram
histogram[0].set_hatch('x')
histogram[1].set_hatch('---')
histogram[2].set_hatch('.')
histogram[3].set_hatch('/')
```

```python
# CONTOH (mengganti warna pada diagram sebelumnya dr pink menjadi turquoise )

# Mendefinisikan data untuk kumpulan data yang akan diproses pada diagram batang
# Penyanyi sebagai label pada sumbu x, dan penggemar sebagai nilai pada sumbu y
penyanyi=['Armand','Ariel','Once','Ari Lasso']
penggemar=[99,49,29,150]

# Mengatur ukuran dari figure yang akan menampilkan histogram.
plt.figure(figsize=(8,4))

# Membuat histogram pada objek hist dengan memasukkan argumen penyanyi sebagai label pada sumbu x dan
# penggemar sebagai nilai pada sumbu y.
# Memberi warna turquoise pada batang
hist=plt.bar(penyanyi,penggemar,color='turquoise')
pola=['x','.','--','/']

for i in [0,1,2,3]:
    hist[i].set_hatch(pola[i])
    
plt.show()
```

## Contoh Pie Chart (Diagram Lingkaran)

```python
import matplotlib.pyplot as plt

# mMmbuat objek figure 
fig = plt.figure()

# Membuat objek axes menggunakan fig.add_axes() dan mengatur nilai [0,0,1,1] sebagai parameter.
ax = fig.add_axes([0,0,1,1])

# Mengatur skala sumbu x dan y
ax.axis('equal')

# Mendefinisikan data untuk kumpulan data yang akan diproses yaitu MataKuliah sebagai label dan Nilai sebagai nilai.
MataKuliah = ['Matematika', 'Fisika', 'Kimia', 'Komputer', 'Bahasa']
Nilai = [80,90,65,79,82]

# Memanggil fungsi pie
ax.pie(Nilai,labels=MataKuliah,autopct='%1.2f%%')

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Mendefinisikan data untuk sumbu x dan y
buah = ['apel', 'bluberi', 'ceri', 'jeruk']
jumlah = [40, 100, 30, 55]

# Mendefinisikan label dan warna untuk setiap bar dengan bar_labels dan bar_colors.
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(buah, jumlah, label=bar_labels, color=bar_colors)

# Mengatur label sumbu-y 
ax.set_ylabel('Permintaan buah')

# Mengatur judul plot 
ax.set_title('Permintaan Buah berdasarkan jenis dan warna')
ax.legend(title='Warna buah')

#  Menampilkan plot 
plt.show()
```

### Pembuatan Scatter Plot (Plot Sebaran) dengan Matplotlib

```python
#SCATTER PLOT
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
tugas = [80,90,65,79,82]
uts = [70,80,85,59,92]
uas = [82,92,87,77,84]

# hubungan tugas dengan uts
ax.scatter(tugas,uts,color='red')
# hubungan tugas dengan uas
ax.scatter(tugas,uas,color='blue')

# set label X dan Y
ax.set_xlabel('Tugas')
ax.set_ylabel('Nilai')
ax.set_title('Scatter Plot')

# tampilkan
plt.show()
```
